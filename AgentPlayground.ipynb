```
import warnings
import logging
import sys
import os

warnings.filterwarnings("ignore", message=".*datetime.datetime.utcnow.*", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)

logging.getLogger("httpx").setLevel(logging.WARNING)

# SETUP: Initialize the AI client directly
# Replace 'your-key-here' with your actual OpenAI API key.
client = OpenAI(api_key="sk-your-key-here")

def run_agent(user_question):
    print(f"User Question: {user_question}")

    # рџџў STEP 1: SEARCH
    # We use DuckDuckGo to grab the latest information from the web.
    print(f"рџ”Ћ Searching the web...")
    with DDGS() as ddgs:
        # Get the top 3 snippets from the internet
        results = [r['body'] for r in ddgs.text(user_question, max_results=3)]
        web_context = "\n\n".join(results) if results else "No information found."

    # рџџЎ STEP 2: THINK
    # We feed the search results into the AI so it can process the facts.
    print("рџ§  Thinking and synthesizing...")
    messages = [
        {"role": "system", "content": "Use the provided web context to answer the question."},
        {"role": "user", "content": f"Web Context:\n{web_context}\n\nQuestion: {user_question}"}
    ]

    # рџ”µ STEP 3: ANSWER
    # The AI generates a polished, human-friendly response.
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )

    return response.choices[0].message.content

# --- Run the Agent ---
if __name__ == "__main__":
    answer = run_agent("Top 3 AI breakthroughs in February 2022")
    print(f"\nвњЁ AI Response:\n{answer}")
```
