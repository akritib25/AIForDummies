```
import os
from openai import OpenAI
from duckduckgo_search import DDGS

# 1️⃣ Setup OpenAI client
# This is like connecting to the AI brain.
# Replace "your-key-here" with your actual OpenAI API key.
client = OpenAI(api_key="your-key-here")

# 2️⃣ Helper function: simple web search
def simple_search(query):
    """
    Perform a quick web search using DuckDuckGo.
    Returns the first few search results as a string.
    """
    with DDGS() as ddgs:  # Open a DuckDuckGo session
        # Grab the text from the top 3 results
        results = [r['body'] for r in ddgs.text(query, max_results=3)]
        return "\n".join(results)  # Combine results into one string

# 3️⃣ Agent function: the AI assistant
def agent(user_input):
    # Print what the user asked
    print(f"User: {user_input}")

    # 🧠 Step 1: Ask the AI to turn your question into a search query
    prompt = f"Convert this to a search engine query: {user_input}"
    search_query = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    ).choices[0].message.content

    # 👀 Step 2: Use that search query to fetch info from the web
    print(f"Searching for: {search_query}...")
    context = simple_search(search_query)

    # ✨ Step 3: Combine the search results and ask AI to give a final answer
    final_prompt = f"Based on this info:\n{context}\n\nAnswer: {user_input}"
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": final_prompt}]
    )

    # Return the AI's final answer
    return response.choices[0].message.content

# 4️⃣ Run the agent with a sample question
print(f"\nAI: {agent('Top 3 AI breakthroughs in February 2026')}")
```
